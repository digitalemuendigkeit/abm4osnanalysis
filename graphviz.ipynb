```python
import networkx as nx
import os.path
import matplotlib.pyplot as plt
import pandas as pd
```

```python
g = nx.read_gml(os.path.join('dataexchange', '_', 'graph_10.gml'), label=None)
```

```python
d = pd.read_csv(os.path.join('dataexchange', '_', 'agent_log.csv'))
```

```python
colors = ['red' if opinion < 0 else 'blue' for opinion in list(d[d['TickNr'] == 100]['Opinion'])]
sizes = [followers * 5 for followers in list(d[d['TickNr'] == 100]['Indegree'])]
```

```python
plt.figure(figsize=(12, 12))
nx.draw_kamada_kawai(g, node_size=sizes, alpha=0.3, arrows=True, node_color=colors)
```
